```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_bgr = cv2.imread('data/lena.jpg')
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(img_bgr)
plt.subplot(1, 2, 2)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.cornerHarris(img_gray, 2, 3, 0.04)
```

```python
plt.imshow(corners, cmap='gray')
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp = sift.detect(img_bgr)
```

```python
import numpy as np
img_kp = np.zeros_like(img_bgr)
img_kp = cv2.drawKeypoints(img_rgb, kp, img_kp, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)#参数：原图，关键点，输出，绘图功能的标识设置
```

```python
plt.imshow(img_kp)
```
