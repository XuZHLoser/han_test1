```python
 x = np.linspace(0, 10, 100)
```

```python
y_true = np.sin(x) + np.random.rand(x.size) - 0.5
```

```python
y_pred = np.sin(x)
```

```python
plt.plot(x, y_pred, linewidth=4, label='model')
plt.scatter(x, y_true, color='r', label='data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
```
